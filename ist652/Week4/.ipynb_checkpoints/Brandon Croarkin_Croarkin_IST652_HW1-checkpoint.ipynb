### Data Cleaning 

The data exploration for this data was fairly brief, since there were no missing data in the file. Then I looked at the shape and column names to view how big the file is and have an easy reference to column names. With this information I was able to spot a few fields that could be dropped or edited for analysis. 

There were two columns with a Row ID that were both repetitive and unneeded due to the automatic indexing of columns done by pandas, so I removed these columns. Next, I made an additional column that stored the zip code data in one column to make further analysis of the zip codes easier. 

```python
###READING IN THE DATA
df = pd.read_csv('donors_data.csv')
```

```python
###BASIC DATA EXPLORATION
df.head()
```

```python
#Check shape of df
df.shape
```

```python
#Explore column names
df.columns
```

```python
###Check for Missing Data
df.isnull().values.any()
```

```python
###DATA CLEANING
df = df.drop(columns = ['Row Id.']) #drop unneeded column
df = df.drop(columns = ['Row Id']) #drop unneeded column
zip_code = [] #create a zip code column from the dummy variables
for i in range(len(df)):
    zips = list(df[['zipconvert_2', 'zipconvert_3','zipconvert_4','zipconvert_5']].iloc[i])
    if np.sum(zips) == 0:
        zip_code.append(1)
    else:
        zip_code.append(zips.index(1) + 2)
df['zip_codes'] = zip_code
```

```python
df.head()
```

### DATA EXPLORATION

```python
##Question 1: For each wealth type, what is the average home values of all the
##donors of that type?

#isolate the variables needed for the question
df_Q1 = df[['WEALTH', 'homeowner dummy', 'HV']]
#subset to just homeowners
df_Q1_homeowners = df_Q1[df_Q1['homeowner dummy'] ==  1][['WEALTH', 'HV']]

#pivot on the WEALTH variable 
Q1_homeowners = pd.pivot_table(df_Q1_homeowners, index = ['WEALTH'])
Q1 = pd.pivot_table(df_Q1, index = ['WEALTH'])
Q1
```

```python
#plot 
plt.plot(Q1_homeowners.index, Q1_homeowners['HV'])
plt.title("Home Values by Wealth Level")
plt.xlabel("Wealth Rating")
plt.ylabel("Average Home Value in $ thousands")
```

The analysis above shows the home values vs. the wealth rating of the individuals. In the first table shown above, there is also a column that shows the percentage of individuals that do own a home. There is a clear positive relationship between wealth and home-ownership with more wealth leading to high home-ownership. 

Once the data was filtered to just include home-owners, the relationship between home value and wealth was very easily evident. The plot above clearly demonstrates this trend and has an interesting spike at the highest wealth level. 

To reach these analyses, the data was first filtered to just the variables of interest and then pivoted to get a table with summations. 

```python
##Question 2: Compare donors in the various zip codes with various
#amounts of giving.
df_Q2 = df[['zip_codes', 'INCOME', 'RAMNTALL']]
pd.pivot_table(df_Q2, index = ['zip_codes'])
```

The analysis above shows how different zip codes differ in their donation levels. It also gives interesting information on the differing socio-economic levels in different zip-codes. Zip-code 1 has the lowest income and lowest amount of lifetime giving, so there seems to be a clear association between the two. However, while the income and lifetime donations are higher in the remaining zip-codes, there does not appear to be ny noticeable differences in lifetime donations. 

To reach these conclusions, the data was first cleaned to create a column with all the zip-code information contained in it so the data could be pivoted on this variable. 

```python
##Question 3: Compare the number of months since the last donation to the 
#last donation amounts.
df_Q3 = df[['totalmonths', 'LASTGIFT']]
pivot_Q3 = pd.pivot_table(df_Q3, index = ['totalmonths'])
pivot_Q3
```

```python
#plot
plt.plot(pivot_Q3.index, pivot_Q3['LASTGIFT'])
plt.title("Number of Months Since Last Donation vs. Size of Last Gift")
plt.xlabel("Number of Months")
plt.ylabel("Size of Last Gift")
```

The final analysis looked at the number of months since the last donation and the size of the gift. Although there is a lot of volatility shown in the graph, there does appear to be an overall trend that those who donated longer ago, donated more. This is an interesting finding and may mean that donations are fairly constant among individuals despite the frequency of the donations, since those who wait longer give more. 

Again, to reach this finding the data was filtered to the variables of interest and then pivoted to get summations.
