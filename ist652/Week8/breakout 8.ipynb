```python
import pandas as pd
import numpy as np
import csv
import json
import requests
```

```python
url = 'https://api.iextrading.com/1.0/stock/{0}/quote'
```

```python
#read csv
symbols = []
shares = []
costbasis = []
    
with open('Portfolio.csv') as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        symbol = row[0]
        share = row[1]
        cost = row[2]

        symbols.append(symbol)
        shares.append(share)
        costbasis.append(cost)

    print(symbols)
    print(shares)
    print(costbasis)
```

```python
symbols = symbols[1:10]
```

```python
shares = shares[1:10]
```

```python
costbasis = costbasis[1:10]
```

```python
url_list = []

for x in symbols:
    newurl = url.replace('{0}', x)
    url_list.append(newurl)
```

```python
print(url_list)
```

```python
#create json file from url_list
with open('result.json', 'w') as fp:
    json.dump([requests.get(url).json() for url in url_list], fp, indent=2)
```
